## Outlier Detection

```python
df2.head()
```

```python
df2["Days for shipping (real)"].value_counts()
```

```python
df2["Order Item Discount"].value_counts()
```

```python
Q1= df2["Order Item Discount"].quantile(0.25)
Q3= df2["Order Item Discount"].quantile(0.75)

IQR=Q3-Q1
min_bound=Q1-(1.5*IQR)
max_bound=Q3+(1.5*IQR)

outliers=df2[(df2["Order Item Discount"] <min_bound) | (df2["Order Item Discount"] > max_bound)]
number_of_outliers=outliers.shape[0]
number_of_outliers
```

```python
Q1= df2["Order Item Product Price"].quantile(0.25)
Q3= df2["Order Item Product Price"].quantile(0.75)

IQR=Q3-Q1
min_bound=Q1-(1.5*IQR)
max_bound=Q3+(1.5*IQR)

outliers=df2[(df2["Order Item Product Price"] <min_bound) | (df2["Order Item Product Price"] > max_bound)]
number_of_outliers=outliers.shape[0]
number_of_outliers
```

## Data Tranformation 


```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
data1["log_Order Item Discount"] = np.log(data1["Order Item Discount"]+1) ## log transformation, log(0) is undefined, So adding +1
data1["log_Order Item Product Price"] = np.log(data1["Order Item Product Price"]+1)
data1["log_Order Item Profit Ratio"] = np.log(data1["Order Item Profit Ratio"]+1)
```
